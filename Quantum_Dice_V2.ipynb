```python
import qiskit
```

```python
def dice():
    while True:
        try:
            sides = input("How many sides do you want your dice to have? (6, 10, 12, 16, 20, 24, 28, 32)")
            if int(sides) in [6, 10, 12, 16, 20, 24, 28, 32]:
                return int(sides)
                break
            else:
                print("Do did not section one of the options. Please try again\n")
                continue
        except:
            print("You did not enter a number. Please try again\n")
            continue

def shots ():
    while True:
        try:
            shots = input("How many numbers do you want the quantum computer to generate? ")
            if int(shots) > 0:
                return int(shots)
                break
            else:
                print("Do did not enter a number more than 0. Please try again/n")
                continue
        except:
            print("You did not enter a number. Please try again/n")
            continue   
        

# UI/UX
sides = dice()
shots = shots()
print("\nRolling a dice with " + str(sides) + " sides " + str(shots) + " times")


# Creating the QC circuit
qc = qiskit.QuantumCircuit(5, 5)
for i in range(5):
    qc.h([i])
    qc.measure([i], [i])
# print(qc)


# Running the QC circuit + print bitstrings
simulator = qiskit.Aer.get_backend("qasm_simulator")
job = qiskit.execute(qc, simulator, shots=(shots*32), memory=True)
results = job.result()
memory = results.get_memory()
# print("\nBitstring results are:")
# print(memory)


# Convert bitstrings to numbers
numbers = []
for i in range(len(memory)):
    number = int(memory[i], 2) + 1
    numbers.append(number)
# print(numbers)


# Find all the numbers that are within range of the number of sides
formated_numbers = []
for i in range(len(numbers)):
    if numbers[i] <= sides:
        formated_numbers.append(numbers[i])
# print(formated_numbers)
    
    
# Print out numbers according to number of shots
print("\nYour generated numbers are:")
for i in range(shots):
    print(formated_numbers[i])
    
```
