```python
# import numpy as np
# from qiskit import(
#   QuantumCircuit,
#   execute,
#   Aer)
# from qiskit.visualization import plot_histogram

import qiskit
import math
```

```python
def dice ():
    while True:
        sides = input("Declare the amount of sides you want in your dice (2, 4, 8, 16, or 32): ")
        try:
            sides = int(sides)
            if sides in [2, 4, 8, 16, 32]:
                return sides
                break
            else:
                print("You did not enter a proper option\n")
                continue
        except:
            print("You did not enter a number\n")
            continue

            
def shots ():
    while True:
        try:
            shots = input("How many numbers do you want the quantum computer to generate? ")
            if int(shots) > 0:
                return int(shots)
                break
            else:
                print("Do did not enter a number more than 0. Please try again")
                continue
        except:
            print("You did not enter a number. Please try again")
            continue            
            
            
# UI/UX
sides = dice()
shots = shots()
print("\nThe dice will have " + str(sides) + " sides")
number_of_qubits = round(math.log2(sides))
print("The quantum computer will being using " + str(number_of_qubits) + " qubits and generating " + str(shots) + " numbers")


# Build QC circuit
circuit = qiskit.QuantumCircuit(number_of_qubits, number_of_qubits)
for i in range(number_of_qubits):
    circuit.h([i])
    circuit.measure([i], [i])
print(circuit)


# Run QC
simulator = qiskit.Aer.get_backend("qasm_simulator")
job = qiskit.execute(circuit, simulator, shots=shots, memory=True)
results = job.result()
memory = results.get_memory()
# counts = results.get_counts()


# Print raw bitstrings
print("\nThe raw bitstrings are:")
print(memory)


# Convert the bitstrings to numbers + 1 (due to 0 index)
print("\nYour generated numbers are:")
for i in range(len(memory)):
    print((int(memory[i], 2)) + 1)
```
